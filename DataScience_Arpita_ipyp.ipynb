```python
#Using Pyhton Lists
import numpy as np
import time
size=1_000_000
l1=list(range(size))
l2=list(range(size))
start=time.time()
add=[x+y for x , y in zip(l1,l2)]
end=time.time()
print(add[0:10])
print(end-start)
```

```python
#Using Numpy Arrays
import numpy as np
import time
size=1_000_000
l1=np.array(list(range(size)))
l2=np.array(list(range(size)))
start=time.time()
add=l1+l2
end=time.time()
print(add[0:10])
print(end-start)
```

Broadcasting in NumPy

```python
#Looping over arrays in Python
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
result = []
for num in arr:
    result.append(num ** 2)
print(result)
```

```python
#Vectorized Operation
arr = np.array([1, 2, 3, 4, 5])
result = arr ** 2  # Vectorized operation
print(result)
```

```python
#Broadcasting with Scalar
result + 10
```

```python
#Broadcasting with arrays of different shapes
arr1 = np.array([1, 2, 3])
arr2 = np.array([10, 20, 30])
result = arr1 + arr2
print(result)
```

```python
#Broadcasting a 2D array and a 1D Array
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([1, 2, 3])
result = arr1 + arr2
print(result)
```

```python
#Hands-on: Applying Broadcasting to Real-World Scenarios
#Normalizing Data Using Broadcasting
# Simulating a dataset (5 samples, 3 features)
data = np.array([[10, 20, 30],
                 [15, 25, 35],
                 [20, 30, 40],
                 [25, 35, 45],
                 [30, 40, 50]])
mean = data.mean(axis=0)
std = data.std(axis=0)
normalized_data = (data - mean) / std
print(normalized_data)
```
