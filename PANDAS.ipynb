```python
# MELT
import pandas as pd
# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}
df = pd.DataFrame(data)
print(df)
```

```python
df.melt(id_vars=["Name"], value_vars=["Math", "Science", "English"], var_name="Subject", value_name="Score")
```

```python
df_melted = df.melt(id_vars=["Name"], value_vars=["Math", "Science", "English"], var_name="Subject", value_name="Score")
df_melted.pivot(index="Name", columns="Subject", values="Score")
```

```python
# .groupby()nfunction
import pandas as pd
df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})
df
```

```python
df.groupby("Department")["Salary"].mean()
```

```python
type(df.groupby("Department")["Salary"].mean())
```

```python
# common aggregation function
df.groupby("Team")["Salary"].sum()      # Total score
```

```python
df.groupby("Team")["Salary"].count()  # how many entries
```

```python
df.groupby("Team")["Salary"].min()
```

```python
df.groupby("Team")["Salary"].max()
```

```python
df.groupby(["Team", "Gender"])["Salary"].mean()   # to group by multiple columns
```

```python
#  .agg and .aggregate are exactly the same
df.groupby("Team")["Salary"].agg(["mean", "max", "min"])
```

```python
type(df.groupby("Team")["Salary"].agg(["mean", "max", "min"]))
```

```python
df.groupby("Team")["Salary"].agg(
    avg_score="mean",
    high_score="max"
)
```

```python
# .transform() Example:
df["Team Avg"] = df.groupby("Team")["Salary"].transform("mean")
df
```

```python
# .filter() Example:
df.groupby("Team").filter(lambda x: x["Salary"].mean() > 85)
```
